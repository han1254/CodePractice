```python
df = pd.read_csv('./maotai.csv')
df
```

```python
# 删除指定列
df.drop(labels='Unnamed: 0', axis=1, inplace=True)# axis=0表示行
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
# 将date作为源数据的行索引
df.set_index('date', inplace=True)
```

```python
df
```

```python
# 输出收盘价比开盘价上涨3%的日期
# （收盘-开盘）/开盘 > 0.03
((df['close'] - df['open'])/df['open']) > 0.03
```

```python
# 获取了符合条件的行数据
df.loc[((df['close'] - df['open'])/df['open']) > 0.03]
```

```python
df.loc[((df['close'] - df['open'])/df['open']) > 0.03].index
```

```python
# 开盘比前日收盘跌幅超过2%
((df['open'] - df['close'].shift(1)))/df['close'].shift(1) < -0.02
```

```python
df.loc[((df['open'] - df['close'].shift(1)))/df['close'].shift(1) < -0.02].index
```

```python
# 从10年1月1日开始，每月第一个交易日，每年最后一个交易日卖出
new_df = df['2010':'2021']
new_df
```

+ 每月买入

    + 每年12次买入，买入1200股，以当期的开盘价进行股票的买卖
    + 一个完整的年，需要卖出1200支股票，以开盘价为单价
    + 2021年只可以买，不可以卖，但是手里剩余的股票需要计算到总收益中

```python
# 对数据进行指定取样
df_monthly = new_df.resample('M').last()
df_monthly
```

```python
cost = df_monthly['open'].sum()*100
```

```python
df_yearsly = new_df.resample('A').last()[:-1]
df_yearsly
recv = df_yearsly['open'][-1]
```

```python
last_price = 8 * 100 * new_df['open']
last_price + recv - cost
```
