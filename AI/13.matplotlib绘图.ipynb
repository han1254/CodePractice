### 直方图
- 是一个特殊的柱状图，又叫做密度图
- plt.hist()的参数
    + bins  
    可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
    + normed  
    如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
    + color  
    指定直方图的颜色。可以是单一颜色值或颜色的序列。如果指定了多个数据集合,例如DataFrame对象，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
    + orientation  
    通过设置orientation为horizontal创建水平直方图。默认值为vertical

### 饼图
- pie()，饼图也只有一个参数x
- 饼图适合展示各部分占总体的比例，条形图适合比较各部分的大小

```python
arr=[11,22,31,15]
plt.pie(arr)
```

```python
arr=[0.2,0.3,0.1]
plt.pie(arr)
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'])
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3)
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3,autopct='%.6f%%')
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3,shadow=True,explode=[0.2,0.3,0.2,0.4])
```

### 散点图scatter()
- 因变量随自变量而变化的大致趋势

- 箱型图
    - 是一种用作显示一组数据分散情况的统计图。因形状如箱子而得名。在各种领域也经常被使用。
    

- 参数


```python
all_data=[np.random.normal(0,std,100) for std in range(1,4)]
figure,axes=plt.subplots() #得到画板、轴
axes.boxplot(all_data,patch_artist=True) #描点上色
plt.show() #展示
```
